```python
import os
import cv2
import numpy as np
```

```python
path_train = "../Train"
number_img = 0
sum_width = 0
sum_height = 0
for dir, subdirs, files in os.walk(path_train):
    for subdir in subdirs:
        path_now = os.path.join(path_train, subdir)
        print(path_now)
        for file in os.listdir(path_now):
            number_img += 1
            print(os.path.join(path_now, file))
            img = cv2.imread(os.path.join(path_now, file))
            height = img.shape[0]
            width = img.shape[1]
            sum_width += width
            sum_height += height
            print(height, width)
```

```python
print("number of image : ", number_img)
print("MEAN_WIDTH : ", sum_width/number_img)
print("MEAN_HEIGHT : ", sum_height/number_img)
```
