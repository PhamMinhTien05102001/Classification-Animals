```python
from os import path
import numpy as np
import cv2
import glob
import matplotlib.pyplot as plt
```

```python
labels = ['Butterfly', 'Chicken', 'Horse', 'Lizard', 'Spider']
sumWidth = 0
sumHeight = 0
sumWidth1 = 0
sumHeight1 = 0
sumWidth2 = 0
sumHeight2 = 0
sumWidth3 = 0
sumHeight3 = 0
sumWidth4 = 0
sumHeight4 = 0
#D:\My Dataset\Classification\Butterfly
path = glob.glob('..\Train/Butterfly\*')
path1 = glob.glob('..\Train\Chicken\*')
path2 = glob.glob('..\Train\Horse\*')
path3 = glob.glob('..\Train\Lizard\*')
path4 = glob.glob('..\Train\Spider\*')
cv_img = []
cv_img1 = []
cv_img2 = []
cv_img3 = []
cv_img4 = []
```

```python
for img in path:
    n = cv2.imread(img)
    cv_img.append(n)
    sumWidth += n.shape[1]
    sumHeight += n.shape[0]
sumWidth = sumWidth//len(cv_img)
sumHeight = sumHeight//len(cv_img)
print (sumWidth)
print(sumHeight)
```

```python
for img in path1:
    n1 = cv2.imread(img)
    cv_img1.append(n1)
    sumWidth1 += n1.shape[1]
    sumHeight1 += n1.shape[0]
sumWidth1 = sumWidth1//len(cv_img1)
sumHeight1 = sumHeight1//len(cv_img1)
print (sumWidth1)
print(sumHeight1)
```

```python
for img in path2:
    n2 = cv2.imread(img)
    cv_img2.append(n2)
    sumWidth2 += n2.shape[1]
    sumHeight2 += n2.shape[0]
sumWidth2 = sumWidth2//len(cv_img2)
sumHeight2 = sumHeight2//len(cv_img2)
print (sumWidth2)
print(sumHeight2)
```

```python
for img in path3:
    n3 = cv2.imread(img)
    cv_img3.append(n3)
    sumWidth3 += n3.shape[1]
    sumHeight3 += n3.shape[0]
sumWidth3 = sumWidth3//len(cv_img3)
sumHeight3 = sumHeight3//len(cv_img3)
print (sumWidth3)
print(sumHeight3)
```

```python
for img in path4:
    n4 = cv2.imread(img)
    cv_img4.append(n4)
    sumWidth4 += n4.shape[1]
    sumHeight4 += n4.shape[0]
sumWidth4 = sumWidth4//len(cv_img4)
sumHeight4 = sumHeight4//len(cv_img4)
print (sumWidth4)
print(sumHeight4)
```

```python
"""Ve so do TB"""
width_means = [sumWidth, sumWidth1, sumWidth2, sumWidth3, sumWidth4]
height_means = [sumHeight, sumHeight1,sumHeight2, sumHeight3, sumHeight4]

x = np.arange(len(labels))  # lập list thứ tự các lable
#print(x)
width = 0.35  # the width of the bars

fig, ax = plt.subplots(1, 1, figsize=(20, 10))

rects1 = ax.bar(x - width/2, width_means, width,color = 'blue', label='Width') 
rects2 = ax.bar(x + width/2, height_means, width,color = 'orange', label='Height')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Size')
ax.set_title('Mean Size of pictures')
ax.set_xlabel('Animals')
ax.set_xticks(x)# đánh dấu nhãn theo list x
ax.set_xticklabels(labels)#đánh dấu các trục bằng tên list labels
ax.legend() #hiện chú giải cho sơ đồ theo 2 rect

ax.bar_label(rects1, padding=3) #Đánh dấu số lượng ảnh trên cột với khoảng cách với cột = 3
ax.bar_label(rects2, padding=3)

fig.tight_layout()
plt.show()
```
